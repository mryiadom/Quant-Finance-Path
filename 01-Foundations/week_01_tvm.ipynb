```python
'''
If (1 + r) = 1.10 (which is a 10% gain), and we square it for two years (1.10^2), we get 1.21.If you started with $1,000, what would your total amount (FV) be after 2 years using that multiplier of 1.21?
'''

pv = 1000
n = 2
r = 10/100

fv = round(pv * pow((1 + r), n), 2)
print(f'The total amount after {n} years would be {fv}')
```

```python
# we want to define a function that will handle fv

def calculate_future_value(pv, n, r):
  fv = round(pv * pow((1 + r), n), 2)
  return fv

initial_investment = 1000
years = 2
interest_rate = 0.1

result = calculate_future_value(initial_investment, years, interest_rate)
print(f'After {years} years, your investment will be worth £{result}')
```

```python
# how to calculate the reverse
'''
In Quant Finance, we often do the opposite. We know we need $2,000 in 5 years, and we want to know how much we need to invest today to get there. This is called Discounting.
we need to rearrange to find PV

pv = fv/(1+r)^n
'''
def calculate_initial_investment(fv, n, r):
  pv = round(fv / pow((1 + r), n), 2)
  return pv

future_value = 2000
years = 5
interest_rate = 0.07

result = calculate_initial_investment(future_value, years, interest_rate)
print(f'If you want £{future_value} in {years} years, you need to invest £{result}')
```
